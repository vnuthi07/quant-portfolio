### **Problem 1: Username Construction**

Write a program that:

- Prompts the user for a first name and last name
- Constructs a username as:
    - first 3 letters of first name
    - first 3 letters of last name
    - followed by a random integer between 100 and 999

Requirements:
- You must define a function:
`generate_username(first_name, last_name)`

You must:
- Use slicing ([:]) at least once
- Use .lower()
- Use random.randint

You may NOT:
- Hardcode any part of the username
- Use global variables inside your function

Edge cases:
- Names shorter than 3 characters must still work correctly

```python
import random

first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")

def generate_username(first_name, last_name):
    if len(first_name) < 3:
        first_part = first_name.lower()
    else:
        first_part = first_name[0:3].lower()
    if len(last_name) < 3:
        second_part = last_name.lower()
    else:
        second_part = last_name[:3].lower()

    third_part = str(random.randint(100, 999))
    return first_part + second_part + third_part

print("Here is your username: " + generate_username(first_name, last_name))
```


### **Problem 2: Password Validation**

Write a program that:
- Prompts the user for a password
- Returns `"Weak"`, `"Medium"`, or `"Strong"`

Requirements:
- You must define:
`check_password_strength(password)`

You must:
- Use a loop to scan the string
- Use at least 2 Boolean flags (e.g., has_digit, has_letter)
- Use if/elif/else

You may NOT:
- Use .isalnum(), .isdigit(), .isalpha()
- Use regular expressions
- Hardcode answers

```python
password = input("Enter a password: ")

def check_password_strength(password):
    has_letter = False
    has_digit = False
    has_special = False
    if len(password) < 6:
        return "weak"
    else:
        for char in password:
            if ('a' <= char <= 'z'):
                has_letter = True
            elif ('0' <= char <= '9'):
                has_digit = True
            else:
                has_special = True
        
    if has_letter and has_digit and has_special:
        return "strong"
                
    elif has_letter and has_digit:
        return "medium"
    else:
        return "weak"
            

print("Your password is " + check_password_strength(password))
```


### **Problem 3: Palindrome Analyzer**

Write a program that:
- Prompts the user for a string

Outputs:
1. Whether it is a palindrome (ignore case)
2. The reversed string
3. The number of vowels

Requirements:

You must:
- Use slicing (`[::-1]`) for reversing
- Use a loop to count vowels
- Use at least one helper function

You may NOT:
- Use `.count()`
- Use any built-in palindrome shortcuts

```python
word = input("Enter your word: ")
def palindrome_analyzer (word):
    def check_for_palindrome(word):
        reversed_word = word[::-1]
        if word.lower() == reversed_word.lower():
            return ("Your word \"" + word + "\" is a palindrome")
        else:
            return ("Your word \"" + word + "\" is not a palindrome")
    
    def vowel_counter(word):
        vowel_count = 0
        vowels = ('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')
        for char in word:
            if char in vowels:
                vowel_count+=1
        return str(vowel_count)

    return (check_for_palindrome(word) + ", and \"" + word + "\" has " + vowel_counter(word) + " vowels.")
    


print(palindrome_analyzer(word))
```

### **Problem 4: Number Guessing Game** 

Write a program that:
- Generates a random integer between 1 and 100
- Repeatedly prompts the user to guess the number

After each guess:
- Print `"Too high"` or `"Too low"`

When correct:
- Print `"Correct"` and number of attempts

Requirements:

You must:
- Use a `while` loop
- Convert user input to `int`
- Track number of guesses

You may NOT:
- Use `for` loops
- Exit the program early with break unless correct guess

```python
import random

def random_number_guesser():
    num = random.randint(1, 100)
    count = 0
    
    guess = int(input("Enter your number: "))
    while guess != num:
        if guess < num:
            count+=1
            print("Too low")
        elif guess > num:
            count+=1
            print("Too high")
        
        guess = int(input("Enter your number: "))
    count+=1
    return ("Correct! It took you " + str(count) + " guesses.")


print(random_number_guesser())
```

### **Problem 5: Running Grade Analyzer** 

Write a program that:
- Repeatedly asks the user for grades (0–100)
- Stops when the user enters `"done"`

Outputs:
- Average grade
- Highest grade
- Lowest grade

Constraints:
- Must handle at least one valid input
- Do not use lists

Requirements:

You must:
- Use a `while` loop
- Track:
    - sum
    - count
    - max
    - min

You may NOT:
- Use lists or arrays
- Use built-in `max()` or `min()`

```python
def grade_analyzer():
    grade = input("Enter your grade: ")
    grade_count = 0
    grade_sum = 0
    
    while grade != "done" or grade != "Done":
        grade_sum += int(grade)
        grade_count += 1
        grade = input("enter your grade: ")
    
    return("The sum of your grades is " + str(grade_sum) + "; the amount of grades you inputed is " + str(grade_count))

print(grade_analyzer())
```
